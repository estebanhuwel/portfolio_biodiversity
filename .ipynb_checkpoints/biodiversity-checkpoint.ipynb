```python
import pandas as pd
import numpy as np
import csv
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
# We read and create our new dataframes to check what kind of data we have
# species_info.csv : contains data about different species and their conservation status
# observations.csv : holds recorded sightings of different species at several national parks for the past 7 days

df_species = pd.read_csv('species_info.csv')
df_observations = pd.read_csv('observations.csv')

print(df_species.columns)
print(df_observations.columns)

print(df_species.head())
print(df_observations.head())
```

```python
# We want to explore and start explaining our data.
# 1 QUESTION -------------------------------------------------------------------------------------------
# What is the distribuiton of conservation_status for animals?

print(df_species['conservation_status'].unique()) # here we check how many different status we have

# As we can see, there are NaN values, that could it mean, that we have some species that are not in danger
# Because of our previous analysis we are goint to do two things
#1 create a new DF without the NaN values
#2 update that NaN to 'No Information' on the same DF
df_species['conservation_status'] = df_species['conservation_status'].fillna('No Information')
print(len(df_species))
print(df_species.head())

df_species_endangered = df_species[df_species['conservation_status'] != 'No Information']
print(len(df_species_endangered))
print(df_species_endangered.head())

```

```python
# Now we are going to start visualizing the data for all the species:
conservation_counts_all = df_species['conservation_status'].value_counts()

plt.figure(figsize=(10,6))
conservation_counts_all.plot(kind='bar')
plt.title('Number of Species by Conservation Status')
plt.xlabel('Conservation Status')
plt.ylabel('Number of Species')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
plt.clf()
```

```python
# Now we are going to start visualizing the data with only endagered species status
conservation_counts_endangered = df_species_endangered['conservation_status'].value_counts()

plt.figure(figsize=(10,6))
conservation_counts_endangered.plot(kind='bar')
plt.title('Number of Species by Conservation Status')
plt.xlabel('Conservation Status')
plt.ylabel('Number of Species')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
plt.clf()

plt.figure(figsize=(10, 6))
plt.pie(conservation_counts_endangered, labels=conservation_counts_endangered.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Number of Species by Conservation Status')
plt.axis('equal')
plt.show()
plt.clf()
```

```python
# 3 QUESTION -------------------------------------------------------------------------------------------
# Distribution of Species by Category
category_counts = df_species['category'].value_counts()

plt.figure(figsize=(10,6))
category_counts.plot(kind='bar')
plt.title('Number of Species by Category')
plt.xlabel('Category')
plt.ylabel('Number of Species')
plt.xticks(rotation = 45, ha='right')
plt.tight_layout()
plt.show()
plt.clf()

plt.figure(figsize=(10, 6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Number of Species by Conservation Status')
plt.axis('equal')
plt.show()
plt.clf()
```

```python
# 4 QUESTION -------------------------------------------------------------------------------------------
# What is the relationship between Category and Conservation Status
category_conservation_table = pd.crosstab(df_species_endangered['category'], df_species_endangered['conservation_status'])

# Creation of the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(category_conservation_table, annot=True, cmap='RdYlGn_r', fmt='d', annot_kws={"size": 6}, cbar_kws={'label': 'Number of Species'})
plt.title('Relationship between Category and Conservation Status')
plt.xlabel('Conservation Status')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
plt.clf()

category_conservation_table.plot(kind='bar', figsize=(14, 8), colormap='RdYlGn_r')
plt.title('Grouped Bar Chart: Category and Conservation Status')
plt.xlabel('Category')
plt.ylabel('Number of Species')
plt.legend(title='Conservation Status')
plt.tight_layout()
plt.show()
plt.clf()
```

```python
# 5 QUESTION -------------------------------------------------------------------------------------------
# Are certain types of species more likely to be endangered?

# As we can see in the previous graph that we have three kind of species that are mostly to be endangered
# Mammal / Vascular Plants / Birds - but we have to do a more exhaustive analysis in each species
categories = df_species_endangered['category'].unique()
total_categories = len(categories)

# Create subplots
fig, axes = plt.subplots(nrows=total_categories, ncols=1, figsize=(20, 5 * total_categories))

for i, category in enumerate(categories):
    ax = axes[i] if total_categories > 1 else axes # this is to select the correct subplot
    category_data = df_species_endangered[df_species_endangered['category'] == category]['conservation_status'].value_counts()
    sns.barplot(x=category_data.index, y=category_data.values, ax=ax, palette='RdYlGn_r')
    ax.set_title(f'Conservation Status of {category}')
    ax.set_xlabel(f'Conservation Status')
    ax.set_ylabel('Number of Species')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
plt.clf()
    
```

```python
# 6 QUESTION -------------------------------------------------------------------------------------------
# Are there differences between species and their conservation status significant?
conservation_status = df_species_endangered['conservation_status'].unique()
total_conservation_status = len(conservation_status)

print(conservation_status)

# Create subplots
fig, axes = plt.subplots(nrows=total_conservation_status, ncols=1, figsize=(20, 5 * total_categories))

for i, status in enumerate(conservation_status):
    ax = axes[i] if total_categories > 1 else axes # this is to select the correct subplot
    conservation_status_data = df_species_endangered[df_species_endangered['conservation_status'] == status]['category'].value_counts()
    sns.barplot(x=conservation_status_data.index, y=conservation_status_data.values, ax=ax, palette='RdYlGn_r')
    ax.set_title(f'Conservation Status - {status}')
    ax.set_xlabel(f'Conservation Status')
    ax.set_ylabel('Number of Species')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
plt.clf()
```

```python
# 7 QUESTION -------------------------------------------------------------------------------------------
# Which species were spotted the most at each park?

print(df_observations.dtypes)
print(df_observations['observations'].mean())
# print(df_observations)

# we will check if we have duplicated rows:
# duplicated_rows = df_observations[df_observations.duplicated(['scientific_name', 'park_name'], keep=False)]
# print(duplicated_rows)

# I want to check only the duplicated rows
# df_observations_duplicated = df_observations.groupby(['scientific_name', 'park_name']).size().reset_index(name='count')
# exact_duplicates = df_observations_duplicated[df_observations_duplicated['count'] > 1]
# print(exact_duplicates.head())

# We want to sum the observations for our duplicated rows, if the scientific name is the same:
# The as_index parameter make the elements grouped to be a standar DF, without using the grouped columns as index
df_observation_grouped = df_observations.groupby(['scientific_name', 'park_name'], as_index=False).agg({'observations': 'sum'})
# print(df_observation_grouped)

# Now we delete the duplicated rows in our original DF
df_observation_no_duplicated = df_observations.drop_duplicates(subset=['scientific_name', 'park_name'])
# print(df_observation_no_duplicated)

# The idea now is to merge the DF without duplicated with the grouped DF so we keep our origina, but updated
# What we are doing here is we are first doing a merge between 2 df, and creating a new column with a new suffix
# then we make our original observations column to be the one in observations_sum
# we drop the new column created, so we have our original DF, but only with the right amount of observations

df_obs_merged = pd.merge(df_observation_no_duplicated, df_observation_grouped, on=['scientific_name', 'park_name'], suffixes=('','_sum'))
df_obs_merged['observations'] = df_obs_merged['observations_sum']
df_obs_merged = df_obs_merged.drop(columns=['observations_sum'])

# print(df_obs_merged)
mean_df_observatinons = df_observations['observations'].mean()
mean_df_obs_merged = df_obs_merged['observations'].mean()

max_df_observatinons = df_observations['observations'].max()
max_df_obs_merged = df_obs_merged['observations'].max()

min_df_observatinons = df_observations['observations'].min()
min_df_obs_merged = df_obs_merged['observations'].min()

print(f'The mean of our original DF is: {mean_df_observatinons}')
print(f'The mean of our merged DF is: {mean_df_obs_merged}')

print(f'The max of our original DF is: {max_df_observatinons}')
print(f'The max of our merged DF is: {max_df_obs_merged}')

print(f'The min of our original DF is: {min_df_observatinons}')
print(f'The min of our merged DF is: {min_df_obs_merged}')

# print(df_obs_merged[df_obs_merged['observations'] > df_obs_merged['observations'].mean()])
```

```python
# With all the information gathered in the last point now is time to start showing everything in a nice way
# First we are going to see in which park we saw more species:
obs_species_count = df_observation_grouped.groupby('park_name')['scientific_name'].nunique().reset_index()
obs_species_count = obs_species_count.rename(columns={'scientific_name': 'num_species'})

print(obs_species_count)

# Inspect the unique scientific names and park names
unique_species = df_observations['scientific_name'].nunique()
unique_parks = df_observations['park_name'].nunique()

print(f"Unique species count: {unique_species}")
print(f"Unique parks count: {unique_parks}")

# Check how many parks each species appears in
species_park_counts = df_observations.groupby('scientific_name')['park_name'].nunique().reset_index()
species_park_counts = species_park_counts.rename(columns={'park_name': 'num_parks'})
species_park_counts_summary = species_park_counts['num_parks'].value_counts().sort_index()

print(species_park_counts_summary)

# Verify if some species are recorded in all parks
all_parks_species = species_park_counts[species_park_counts['num_parks'] == unique_parks]
print(f"Number of species recorded in all parks: {len(all_parks_species)}")
print(all_parks_species.head())

plt.figure(figsize=(20,5))
plt.bar(obs_species_count['park_name'], obs_species_count['num_species'])
plt.xlabel('Park Name')
plt.ylabel('Total Species Seen')
plt.title('Total Species in each Park')
plt.xticks(rotation = 45, ha='right')
plt.tight_layout()
plt.show()
plt.clf()
```

```python
# Because it looks like an error, now I am going to create the same bar but making a sum of all the observations in each park

def thousands_formatter(x, pos):
    return f'{int(x / 1000)}K'

obs_sum = df_obs_merged.groupby('park_name')['observations'].sum().reset_index()
obs_sum = obs_sum.rename(columns={'observations': 'total_observations'})
print(obs_sum)

plt.figure(figsize=(10,6))
plt.bar(obs_sum['park_name'], obs_sum['total_observations'])
plt.xlabel('Park Name')
plt.ylabel('Total Observations fot all the Species Seen')
plt.title('Total Obs. for all the species in each Park')
plt.xticks(rotation = 45, ha='right')

plt.gca().yaxis.set_major_formatter(FuncFormatter(thousands_formatter))

plt.tight_layout()
plt.show()
plt.clf()

plt.figure(figsize=(8, 8))
plt.pie(obs_sum['total_observations'], labels=obs_sum['park_name'], autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))
plt.title('Total Observations in each Park')
plt.show()
plt.clf()
```

```python
# 8 QUESTION -------------------------------------------------------------------------------------------
# Which category-species were spotted the most at each park?

# To solve this question, we can see that we have a column in common in both DataFrames, so will be great to merge
# The merge will be done to the left side, so we keep the information from the df_observations
df_obs_species = df_obs_merged.merge(df_species[['scientific_name', 'category']], on='scientific_name', how='left')
# df_obs_species = df_obs_species.drop(columns=('scientific_name'))

obs_by_park = df_obs_species.groupby('park_name').size()
print("How many datas we have per Park")
print(obs_by_park)

print()
print(df_obs_species.groupby('park_name').nunique())

# obs_sum = df_obs_merged.groupby('park_name')['observations'].sum().reset_index()

# Because for this case we dont need the scientific name we are going to drop this table
print(df_obs_species.head(3))

# Creation of the graphic to see which category was seen in each park
category_obs_table = pd.crosstab(df_obs_species['park_name'], df_obs_species['category'])
print(category_obs_table)

category_obs_table.plot(kind='bar', figsize=(14,8))

plt.title('Grouped Bar Chart: Category and Conservation Status')
plt.xlabel('Category')
plt.ylabel('Number of Species')
plt.legend(title='Conservation Status')
plt.tight_layout()
plt.show()
plt.clf()
```

```python
# 9 QUESTION -------------------------------------------------------------------------------------------
# Which are the most observed species?

most_observed_species = df_obs_merged.loc[df_obs_merged.groupby('park_name')['observations'].idxmax()]
print("Species most observed in each park:")
print(most_observed_species[['park_name', 'scientific_name', 'observations']])
```

```python
# 10 QUESTION -------------------------------------------------------------------------------------------
# The community wants to see the most observed species in each park, and compare them with how they are seen in the other parks.

scientific_name_list = most_observed_species['scientific_name']
total_scientific_name_list = len(scientific_name_list)

# Create subplots with 2 columns: one for bar plots and one for pie charts
fig, axes = plt.subplots(nrows=total_scientific_name_list, ncols=2, figsize=(20, 5 * total_scientific_name_list))

# Iterate over the most observed species and create bar plots and pie charts for each
for i, names in enumerate(scientific_name_list):
    # Select the correct subplot
    ax_bar = axes[i, 0] if total_scientific_name_list > 1 else axes[0]
    ax_pie = axes[i, 1] if total_scientific_name_list > 1 else axes[1]
    
    # Data for the species
    names_data = df_obs_merged[df_obs_merged['scientific_name'] == names].groupby('park_name')['observations'].sum().reset_index()
    names_data = names_data.sort_values(by='observations', ascending=False)
    
    # Bar plot
    sns.barplot(x='park_name', y='observations', data=names_data, ax=ax_bar, palette='RdYlGn_r')
    ax_bar.set_title(f'Observations of {names}')
    ax_bar.set_xlabel('Park Name')
    ax_bar.set_ylabel('Number of Observations')
    ax_bar.set_xticklabels(ax_bar.get_xticklabels(), rotation=45, ha='right')
    
    # Pie chart
    ax_pie.pie(names_data['observations'], labels=names_data['park_name'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('RdYlGn_r', len(names_data)))
    ax_pie.set_title(f'Percentage of Observations of {names}')

plt.tight_layout()
plt.show()
plt.clf()
```
